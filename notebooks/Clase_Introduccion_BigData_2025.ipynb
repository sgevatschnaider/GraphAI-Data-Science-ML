```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from scipy.optimize import curve_fit
from google.colab import files

# Generar un grafo con una distribución de ley de potencias
n = 100  # Número de nodos
m = 2  # Número de enlaces que se añaden desde un nuevo nodo a nodos existentes
grafo = nx.barabasi_albert_graph(n, m)

# Calcular la distribución de grados
degrees = [nx.degree(grafo, n) for n in grafo]
hist = nx.degree_histogram(grafo)

# Graficar la distribución de grados en escala logarítmica
plt.figure(figsize=(12, 8))
plt.loglog(np.arange(len(hist)), hist, marker='o', linestyle='-', label='Distribución de Grado')
plt.xlabel('Grado (log)')
plt.ylabel('Número de nodos (log)')
plt.title("Distribución de Grado del Grafo (Ley de Potencias)")

# Ajustar la distribución de grados a una función de potencia
def power_law(x, a, b):
    return a * np.power(x, b)

hist = np.array(hist)  # Convertir hist a un array de NumPy
x_data = np.arange(len(hist))[hist > 0]
y_data = hist[hist > 0]
popt, pcov = curve_fit(power_law, x_data, y_data)
a, b = popt

# Graficar la función de potencia ajustada
plt.loglog(x_data, power_law(x_data, a, b), '-', label='Ajuste de Ley de Potencias')
plt.legend()
plt.show()

# Dibujar el grafo con información visual
plt.figure(figsize=(12, 8))
nx.draw(grafo, node_size=[nx.degree(grafo, n) for n in grafo], with_labels=True)
plt.title("Grafo que simula la estructura de Internet (Ley de Potencias)")
plt.show()

# Guardar el grafo en un archivo .gml
nx.write_gml(grafo, "red_internet.gml")

# Descargar el archivo a tu computadora
files.download("red_internet.gml")
```

```python
import networkx as nx
import numpy as np
import random as rd
import matplotlib.pyplot as plt
G = nx.barabasi_albert_graph(n = 100, m = 2, seed=10374196, initial_graph = None)
degree_sequence = sorted((d for n, d in G.degree()), reverse=True)
dmax = max(degree_sequence)
fig = plt.figure("Degree of a random graph", figsize=(8, 8))
# Create a gridspec for adding subplots of different sizes
axgrid = fig.add_gridspec(5, 4)
ax0 = fig.add_subplot(axgrid[0:3, :])
Gcc = G.subgraph(sorted(nx.connected_components(G), key=len, reverse=True)[0])
pos = nx.spring_layout(Gcc, seed=10396953)
nx.draw_networkx_nodes(Gcc, pos, ax=ax0, node_size=20)
nx.draw_networkx_edges(Gcc, pos, ax=ax0, alpha=0.4)
ax0.set_title("Scale-Free Network G")
ax0.set_axis_off()
ax1 = fig.add_subplot(axgrid[3:, :2])
ax1.plot(degree_sequence, "b-", marker="o")
ax1.set_title("Gráfico de distirubición de los nodos (2000 nodes)")
ax1.set_ylabel("Grado")
ax1.set_xlabel("Número de nodos")
ax2 = fig.add_subplot(axgrid[3:, 2:])
ax2.bar(*np.unique(degree_sequence, return_counts=True))
ax2.set_title("Grado Histograma")
ax2.set_xlabel("Grado")
ax2.set_ylabel("Número de Nodos")
fig.tight_layout()
plt.show()
```

```python
from queue import Queue
import time

# Crear una cola de trabajos de impresión
cola_de_impresion = Queue()

# Función para simular la adición de trabajos de impresión a la cola
def enviar_trabajo(trabajo):
    cola_de_impresion.put(trabajo)
    print(f"Trabajo {trabajo} añadido a la cola de impresión.")

# Función para simular el procesamiento de trabajos de impresión
def procesar_trabajos():
    while not cola_de_impresion.empty():
        trabajo = cola_de_impresion.get()
        print(f"Procesando {trabajo}")
        # Simular el tiempo de impresión
        time.sleep(2)
        print(f"Trabajo {trabajo} completado.")
        cola_de_impresion.task_done()

# Simular el envío de trabajos de impresión
enviar_trabajo("Trabajo_A")
enviar_trabajo("Trabajo_B")
enviar_trabajo("Trabajo_C")

# Procesar trabajos de impresión en orden FIFO
procesar_trabajos()
```
